```python
import pandas as pd
df = pd.read_excel('S:/Car/ML ready data/car_data_cleaned.xlsx')
df.shape
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor()
rf.fit(df.drop('price_in_lakhs', axis=1), df['price_in_lakhs'])
feature_importances = rf.feature_importances_
important_features = pd.Series(feature_importances, index=df.drop('price_in_lakhs', axis=1).columns)
important_features.sort_values(ascending=False, inplace=True)
print(important_features)
```

```python
print(important_features[important_features > 0.01].index) # important_features.sort_values(ascending=False, inplace=True)
```

```python
#outliers 
from scipy import stats
import numpy as np

# Calculate the Z-scores
z_scores = stats.zscore(df)

# Remove outliers (keeping only those with Z-score < 3 in absolute value)
df = df[(np.abs(z_scores) < 3).all(axis=1)]
```

```python
df
```

```python
df.drop(columns = ['it','Unnamed: 0.1'],axis=1,inplace=True)
```

```python
corr_df = df.corr()
corr_df
```

```python
correlation_with_price = corr_df['price_in_lakhs'].sort_values(ascending=True)
correlation_with_price
```

```python
relevant_features = correlation_with_price[correlation_with_price>0.05].index
relevant_features_2 = correlation_with_price[correlation_with_price<-0.2].index
relevant_features
```

```python
final_features = relevant_features.append(relevant_features_2)
type(final_features)
```

```python
relevant_features_3 = ['ft','ownerNo','modelYear','Insurance Validity','km','transmission','Mileage',
                       'Year of Manufacture','Seats_y','Color','Steering Type','model','Color','Steering Type']
final_features = final_features + relevant_features_3
final_features
```

```python
len(final_features)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

df[final_features] = sc.fit_transform(df[final_features])
```

```python
df
```

```python
df.isna().sum()
```

```python
df['transmission']
```

```python
X = df[final_features]
X.drop('price_in_lakhs',axis=1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, df['price_in_lakhs'], test_size=0.3, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y_test, y_pred)
```
